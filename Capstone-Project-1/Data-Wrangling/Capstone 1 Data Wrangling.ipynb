### 4) Fix zip code column

```python
# convert column to string, then zero-pad on left
```

```python
df['BUYER_ZIP'] = df['BUYER_ZIP'].astype(str)
```

```python
for i, zip_code in enumerate(df['BUYER_ZIP']):
    if len(str(zip_code)) == 4:
        x = '0' + str(zip_code)
        #print(x)
        df['BUYER_ZIP'].values[i] = x
    elif len(str(zip_code)) == 3:
        x = '00' + str(zip_code)
        #print(x)
        df['BUYER_ZIP'].values[i] = x
```

### 5) Same process for date column

```python
df['TRANSACTION_DATE'] = df["TRANSACTION_DATE"].astype(int).astype(str)
```

```python
for i, date in enumerate(df['TRANSACTION_DATE']):
    if len(date) == 7:
        df['TRANSACTION_DATE'].values[i] = '0' + date
```

```python
#We want to convert datetime to a year column to make it easier to join with overdose data and do time analysis
df['date'] = pd.to_datetime(df['TRANSACTION_DATE'],format = '%m%d%Y')
df.head()
df['year'] = df['date'].dt.year
```

### 6) Check for missing values

```python
df_nan = df.loc[df.isnull().any(axis=1)]
df_nan
```

```python
# check if independent cities have a county value
check = df.loc[df.BUYER_CITY == 'BALTIMORE']
print(check)
```

```python
# percent of NaN's is not statistically significant, nor a systematic error
df = df.dropna()
```

### 7) Sort and sum

```python
# Sort by state and year
df.sort_values(['BUYER_STATE', 'year'])
```

```python
# Group by state and year, then sum over quantity
# set as_index to false so that state and year are not set as multi-indexes
sum_state = df.groupby(['BUYER_STATE', 'year'], as_index = False)['QUANTITY'].sum()
```

### 10) Merge dataframes

### 11) Find prescription rate and overdose rate

```python
# create new columns for prescription rate and overdose rate
merge_df['prescription_rate'] = (merge_df['QUANTITY'] / merge_df['population'])* 100000
```

```python
#convert deaths column to type int, then use it to compute overdose rate
merge_df['deaths'] = merge_df['deaths'].astype(int)
merge_df['overdose_rate'] = (merge_df['deaths'] / merge_df['population'])* 100000
```
